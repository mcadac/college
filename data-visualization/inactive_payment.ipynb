```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
df = pd.read_csv('data/inactive_payment_error_tx_total.csv')
df
```

```python
figure(num=None, figsize=(22, 20), dpi=80, facecolor='w', edgecolor='k')

for network in df['red_financiera_pagos'].drop_duplicates():
    print(network)
    networks = df[df['red_financiera_pagos'] == network]
    date = networks['to_char']
    count = networks['count']
    plt.plot(date,count,label=network)
    plt.legend(loc='upper left')
```
